# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

## Hypothesis test

We want to find out if there is a correlation between hospital capacity and readmission rates. Since we are trying to discover a relationship between two variables, we will be using the **Chi Square Test for Independence.**

To conduct the hypothesis test, the following are defined:
* **Null Hypothesis:** There is a correlation between hospital capacity and readmission rates.
* **Alternate Hypothesis:** There is no correlation between hospital capacity and readmission rates.

We assume the null hypothesis to be true to conduct this analysis.

```python
df = clean_hospital_read_df
df
```

We need to classify hospitals based on their capacity. For our purposes, we will divide all the hospitals into 3 types:
* **Big Hospitals:** Number of discharges > 1000
* **Medium Hospitals:** 100 <= Number of discharges <= 1000
* **Small Hospitals:** Number of discharges < 100

```python
df_big =df[df['Number of Discharges'] > 300]
df_small = df[df['Number of Discharges'] <= 300]
```

```python
df_big_less = sum(df_big['Excess Readmission Ratio'] <= 1)
df_big_more = sum(df_big['Excess Readmission Ratio'] > 1)

df_big_less, df_big_more
```

```python
df_small_less = sum(df_small['Excess Readmission Ratio'] <= 1)
df_small_more = sum(df_small['Excess Readmission Ratio'] > 1)

df_small_less, df_small_more
```

```python
len(df_big), len(df_small)
```

## Table of Actual Values

From the values calculated above, we can construct a lookup table that will enable us to visualise the data better.

|   Ratio       | Big    | Small  |  Total |
| ------------- |:------:| ------:|-------:| 
| > 1           | 2641   | 3298   |  5939  |
| <= 1          | 2724   | 2834   |  5558  |
| Total         | 5365   | 6132   |  11497 |


```python
p0 = 5939/11497
p1 = 5558/11497

p0, p1
```

```python
e11 = p0 * 5365
e12 = p0 * 6132
e21 = p1 * 5365
e22 = p1 * 6132


from pprint import pprint

pprint([[e11, e12], [e21, e22]])
```

```python
df = 1
chi2 = (e11-2641)**2/e11 + (e12-3298)**2/e12 + (e21-2724)**2/e21 + (e22-2834)**2/e22 
chi2
```

```python
p = 1-stats.chi2.cdf(chi2, df)
p
```

The chi square value is so large that the p-value is effectively 0. Therefore, we have to reject the null hypothesis and accept the alternate hypothesis. In other words, **there is a correlation between the number of discharges and excess readmission ratio.**

## Statistical Significance and Practical Significance

The result obtained is statistically significant as we have rejected the null hypothesis for statistical significance level $\alpha$=0.01

But more importantly, the result is also practically significant. This is because the value of chi square obtained is so large that the p-value is practically 0. This suggests an extremely strong correlation between hospital capacity and readmission ratio. This, in turn, ensures that the result can be applied to practical decision making.

The suggestions by statistical significance and practical significance do not differ but practical significance ensures a more potent solution to be delivered to the client.
